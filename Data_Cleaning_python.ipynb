# RISK SCORE

```python
df['approval_ratio'] = df['amount_claim_approved'] / df['amount_claim_initiated']
```

```python
df['risk_score'] = (
    df['stay_days'].fillna(0)*0.25 +
    df['claim_delay'].fillna(0)*0.25 +
    df['payment_delay'].fillna(0)*0.2 +
    (1 - df['approval_ratio'].fillna(0))*100*0.3
)
```

# Visualization

```python
total_cases = len(df)

claims_initiated = df['claim_init_date'].notna().sum()
claims_approved = df['claim_approved_date'].notna().sum()
payments_done = df['payment_paid_dt'].notna().sum()

total_amount = df['amount_claim_approved'].sum()
total_loss = (df['amount_claim_initiated'] - df['amount_claim_approved']).sum()

avg_claim_delay = df['claim_delay'].mean()
avg_payment_delay = df['payment_delay'].mean()

print("Total Cases:", total_cases)
print("Claims Approved %:", claims_approved/total_cases*100)
print("Payments Done %:", payments_done/total_cases*100)
print("Total Approved Amount:", total_amount)
print("Revenue Loss:", total_loss)
print("Avg Claim Delay:", avg_claim_delay)
print("Avg Payment Delay:", avg_payment_delay)
```

```python
funnel_values = [
    total_cases,
    claims_initiated,
    claims_approved,
    payments_done
]

labels = ['Total', 'Initiated', 'Approved', 'Paid']

plt.figure()
plt.bar(labels, funnel_values)
plt.title("Claim Funnel")
plt.show()
```
